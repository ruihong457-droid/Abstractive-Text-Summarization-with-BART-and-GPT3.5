```python
from youtube_transcript_api import YouTubeTranscriptApi

def get_transcript(video_id):
    ytt_api = YouTubeTranscriptApi()
    try:
        transcript = ytt_api.fetch(video_id, languages=['en'])
        return " ".join([snippet.text for snippet in transcript])
    except Exception as e:
        print("ERROR: ", e)
        return ""

"""
    GPT-3.5 Summarization
    Note: Since OpenAI no longer provide free API qouta, here we use third-party provider to get the summary.
"""

import openai

API_KEY = ""
client = openai.OpenAI(api_key=API_KEY, base_url = "https://api.agicto.cn/v1")

def summarize_gpt(text):
    prompt = f"""Summarize the following lecture into clear bullet points. Do
            NOT add any information not in the text. Text: {text}"""

    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo-0613",
            messages=[{"role": "user", "content": prompt}],
            temperature=0.2
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print(f"Failed to summarize：{e}")
        return ""
```

```python
video_id = "gN9dlisaQVM"        # This is a TED speech video, which discusses how to turn pressure into friends.
text = get_transcript("gN9dlisaQVM")
print("Transcript:\n", text)

# Use GPT-3.5 to summarize
gpt_summary = summarize_gpt(text)
print("\nSummary:\n", gpt_summary)
```
